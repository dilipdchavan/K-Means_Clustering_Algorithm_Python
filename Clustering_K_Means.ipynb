```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
#set desktop or system path
import sys
sys.path.append('C:\\DCHAVAN\\DSTI-Masters\\DilipPersonalStudy\\Clustering\\Data')
```

```python
heart_data = pd.read_csv('C:\\DCHAVAN\\DSTI-Masters\\DilipPersonalStudy\\Clustering\\Data\\heart.csv')
```

```python
heart_data.describe()
```

```python
X_heart = heart_data.iloc[:,[0,1,2,3,4,5,6,7]]
```

```python
print(X_heart.columns)
```

```python
wcss_heart=[]
for i in range(1,11):
    kmeans_heart = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10)
    y_means_heart = kmeans_heart.fit(X_heart)
    wcss_heart.append(y_means_heart.inertia_)
```

```python
plt.plot(range(1,11), wcss_heart)
plt.show()
```

```python
kmeans_heart = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10)
y_kmeans_heart = kmeans_heart.fit_predict(X_heart)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X_heart.iloc[y_kmeans_heart == 0,7], X_heart.iloc[y_kmeans_heart == 0,4], s= 100, color = 'red', label ='Need to be healthy')
plt.scatter(X_heart.iloc[y_kmeans_heart == 1,7], X_heart.iloc[y_kmeans_heart == 1,4], s= 100, color = 'blue', label ='Healthy')

plt.scatter(kmeans_heart.cluster_centers_[:,7], kmeans_heart.cluster_centers_[:, 4], s= 300, color = 'yellow', label ='Centroid')
plt.title('k-means clustering')
plt.xlabel('Max Heart rate achieved')
plt.ylabel('cholesterol')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X_heart.iloc[y_kmeans_heart == 0,3], X_heart.iloc[y_kmeans_heart == 0,4], s= 100, color = 'red', label ='Need to be healthy')
plt.scatter(X_heart.iloc[y_kmeans_heart == 1,3], X_heart.iloc[y_kmeans_heart == 1,4], s= 100, color = 'blue', label ='Healthy')

plt.scatter(kmeans_heart.cluster_centers_[:,3], kmeans_heart.cluster_centers_[:,4], s= 300, color = 'yellow', label ='Centroid')
plt.title('k-means clustering')
plt.xlabel('Resting blood pressure')
plt.ylabel('cholesterol')
plt.legend()
plt.show()
```
